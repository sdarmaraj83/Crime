```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the parquet file
file_path = 'https://storage.data.gov.my/publicsafety/crime_district.parquet'
data = pd.read_parquet(file_path)

# Remove rows where type is 'all' or state is 'Malaysia'
data = data[(data['type'] != 'all') & (data['state'] != 'Malaysia')]

# Get the summary of the number of crimes by type of crime and state
crime_by_type_and_state = data.pivot_table(index='state', columns='type', values='crimes', aggfunc='sum', fill_value=0)

# Sort the data from highest to lowest total crimes
crime_by_type_and_state['Total Crimes'] = crime_by_type_and_state.sum(axis=1)
crime_by_type_and_state = crime_by_type_and_state.sort_values(by='Total Crimes', ascending=False)

# Plot the summary of the number of crimes by type of crime and state
crime_by_type_and_state.drop(columns='Total Crimes').plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Number of Crimes by Type of Crime and State')
plt.xlabel('State')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Load crime data
file_path = 'https://storage.data.gov.my/publicsafety/crime_district.parquet'
data = pd.read_parquet(file_path)

# Filter rows where type is not 'all' and state is not 'Malaysia'
data = data[(data['type'] != 'all') & (data['state'] != 'Malaysia')]

# Summarize total crimes by state
crime_by_state = data.groupby('state')['crimes'].sum().reset_index()
crime_by_state['state'] = crime_by_state['state'].str.strip().str.lower()  # Normalize state names

# Load the state-level shapefile
shapefile_path = 'Malaysia/gadm41_MYS_1.dbf'  # Update with your path
gdf = gpd.read_file(shapefile_path)

# Normalize state names in the shapefile
gdf['state_name'] = gdf['NAME_1'].str.strip().str.lower()  # Use the `NAME_1` column for state names

# Standardize state names for consistent merging
state_name_mapping = {
    'trengganu': 'terengganu',
    'pulau pinang': 'penang',
    'kuala lumpur': 'w.p. kuala lumpur',
    'labuan': 'w.p. labuan',
    'putrajaya': 'w.p. putrajaya',
}

crime_by_state['state'] = crime_by_state['state'].replace(state_name_mapping)
gdf['state_name'] = gdf['state_name'].replace(state_name_mapping)

# Merge crime data with geographical data
gdf = gdf.merge(crime_by_state, left_on='state_name', right_on='state', how='left')

# Handle missing crime data
gdf['crimes'] = gdf['crimes'].fillna(0)

# Re-project geometries to a projected CRS
gdf = gdf.to_crs(epsg=3395)

# Scale up the geometries to make the states bigger
gdf['geometry'] = gdf['geometry'].scale(xfact=1.2, yfact=1.2, origin='center')


# Plot the geographical map with a larger size
fig, ax = plt.subplots(1, 1, figsize=(20, 15))
gdf.plot(column='crimes', cmap='coolwarm', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add labels for the states
for x, y, label in zip(gdf.geometry.centroid.x, gdf.geometry.centroid.y, gdf['NAME_1']):
    ax.text(x, y, label, fontsize=10, ha='center', color='black')

# Set a fixed aspect ratio
ax.set_aspect('equal')

plt.title('Total Crimes by State in Malaysia', fontsize=16)
plt.axis('off')  # Remove axes for a clean map
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the parquet file
file_path = 'https://storage.data.gov.my/publicsafety/crime_district.parquet'
data = pd.read_parquet(file_path)

# Remove rows where type is 'all' or state is 'Malaysia'
data = data[(data['type'] != 'all') & (data['state'] != 'Malaysia')]

# Get the summary of the total number of crimes by type of crime
crime_summary = data.groupby('type')['crimes'].sum()

# Sort the summary from highest to lowest
crime_summary = crime_summary.sort_values(ascending=False)

# Plot the summary of the total number of crimes by type of crime with color coding
colors = plt.cm.viridis(crime_summary / float(max(crime_summary)))
crime_summary.plot(kind='bar', figsize=(12, 8), color=colors)
plt.title('Total Number of Crimes by Type of Crime in Malaysia from 2016 to 2023')
plt.xlabel('Crime Type')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the parquet file
file_path = 'https://storage.data.gov.my/publicsafety/crime_district.parquet'
data = pd.read_parquet(file_path)

# Remove rows where type is 'all' or state is 'Malaysia'
data = data[(data['type'] != 'all') & (data['state'] != 'Malaysia')]

# Filter data for crime type 'theft_vehicle_motorcycle'
theft_vehicle_motorcycle_data = data[data['type'] == 'theft_vehicle_motorcycle']

# Get the summary of the total number of crimes by state for 'theft_vehicle_motorcycle'
theft_vehicle_motorcycle_summary = theft_vehicle_motorcycle_data.groupby('state')['crimes'].sum()

# Sort the summary from highest to lowest
theft_vehicle_motorcycle_summary = theft_vehicle_motorcycle_summary.sort_values(ascending=False)

# Plot the summary of the total number of crimes by state for 'theft_vehicle_motorcycle'
colors = plt.cm.viridis(theft_vehicle_motorcycle_summary / float(max(theft_vehicle_motorcycle_summary)))
theft_vehicle_motorcycle_summary.plot(kind='bar', figsize=(12, 8), color=colors)
plt.title('Total Number of Theft Vehicle Motorcycle Crimes by State')
plt.xlabel('State')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the parquet file with error handling
file_path = 'https://storage.data.gov.my/publicsafety/crime_district.parquet'
try:
    data = pd.read_parquet(file_path)
except Exception as e:
    print(f"Error loading file: {e}")
    exit()

# Remove rows where type is 'all' or state is 'Malaysia'
data = data[(data['type'] != 'all') & (data['state'] != 'Malaysia')]

# Extract year from the date field
data['year'] = pd.to_datetime(data['date']).dt.year

# Get the summary of the total number of crimes by type of crime and year
crime_summary_type_yearly = data.groupby(['type', 'year'])['crimes'].sum().unstack()

# Sort the summary from highest to lowest based on 2023 numbers
crime_summary_type_yearly = crime_summary_type_yearly.sort_values(by=2023, ascending=False)

# Plot the total number of crimes by type of crime and year with color coding in a column bar chart (not stacked)
crime_summary_type_yearly.plot(kind='bar', figsize=(12, 8), colormap='viridis')
plt.title('Total Number of Crimes by Type of Crime and Year')
plt.xlabel('Crime Type')
plt.ylabel('Number of Crimes')
plt.legend(title='Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the parquet file with error handling
file_path = 'https://storage.data.gov.my/publicsafety/crime_district.parquet'
try:
    data = pd.read_parquet(file_path)
except Exception as e:
    print(f"Error loading file: {e}")
    exit()

# Filter data for rape cases and the years 2020 to 2023
data['year'] = pd.to_datetime(data['date']).dt.year
rape_cases = data[(data['type'] == 'rape') & (data['year'].between(2020, 2023))]

# Remove rows where state is 'Malaysia'
rape_cases = rape_cases[rape_cases['state'] != 'Malaysia']

# Get the summary of rape cases by state for the years 2020 to 2023
rape_cases_summary = rape_cases.groupby(['state', 'year'])['crimes'].sum().unstack()

# Sort the summary from highest to lowest for each year
rape_cases_summary = rape_cases_summary.apply(lambda x: x.sort_values(ascending=False), axis=0)

# Plot the rape cases by state for the years 2020 to 2023
rape_cases_summary.plot(kind='bar', figsize=(12, 8), colormap='viridis')
plt.title('Rape Cases by State for the Years 2020 to 2023')
plt.xlabel('State')
plt.ylabel('Number of Rape Cases')
plt.legend(title='Year')
plt.show()
```
